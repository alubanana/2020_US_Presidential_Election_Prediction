```python
def get_nrc_data():
    nrc = "NRC-emotion-lexicon-wordlevel-alphabetized-v0.92.txt"
    count=0
    emotion_dict=dict()
    with open(nrc,'r') as f:
        all_lines = list()
        for line in f:
            if count < 46:
                count+=1
                continue
            line = line.strip().split('\t')
            if int(line[2]) == 1:
                if emotion_dict.get(line[0]):
                    emotion_dict[line[0]].append(line[1])
                else:
                    emotion_dict[line[0]] = [line[1]]
    return emotion_dict


def emotion_analyzer(text,emotion_dict=get_nrc_data()):
    #Set up the result dictionary
    emotions = {x for y in emotion_dict.values() for x in y}
    emotion_count = dict()
    for emotion in emotions:
        emotion_count[emotion] = 0

    #Analyze the text and normalize by total number of words
    total_words = len(text.split())
    for word in text.split():
        if emotion_dict.get(word):
            for emotion in emotion_dict.get(word):
                emotion_count[emotion] += 1/len(text.split())
    return emotion_count

def comparative_emotion_analyzer(text_tuples,object_name="Restaurant"):
    
    import pandas as pd
    df = pd.DataFrame(columns=[object_name,'Fear', 'Anger', 'Trust','Negative',
                           'Positive','Joy','Disgust','Anticipation',
                           'Sadness','Surprise'],)
    df.set_index(object_name,inplace=True)
    
    output = df    
    for text_tuple in text_tuples:
        text = text_tuple[1] 
        result = emotion_analyzer(text)
        
        df.loc[text_tuple[0]] = [result['fear'], result['anger'], result['trust'],
                  result['negative'],result['positive'],result['joy'],result['disgust'],
                  result['anticipation'],result['sadness'],result['surprise']]
    return output
```

```python
def get_tweets(search_term,consumer_key=consumer_key,consumer_secret=consumer_secret,
               access_token=access_token,access_token_secret=access_token_secret):
    import tweepy
    auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
    auth.set_access_token(access_token, access_token_secret)
    api = tweepy.API(auth)
#    search_term = 'AAPL'
    ## fill in your search query and store your results in a variable
    results = api.search(q = search_term, lang = "en", result_type = "recent", count = 3000)
#    search_term = 'AAPL'
    for i in range(len(results)):
        fname = search_term+'.'+str(len(results)-i)
        with open('tweets/'+fname,'w') as f:
            f.write(results[i]._json['text']+'\n')

```

```python
get_tweets('Donald Trump')
get_tweets('Elizabeth Warren')
get_tweets('Bernie Sanders')
get_tweets('Kamala Harris')
get_tweets("Beto O'Rourke")
get_tweets('Kirsten Gillibrand')
get_tweets('Amy Klobucha')
get_tweets('Cory Booker')
get_tweets('JuliÃ¡n Castro')
```

```python

import nltk
from nltk.corpus import PlaintextCorpusReader
tweets_root = "tweets/"
Trump_files = "Donald Trump.*"
Warren_files = "Elizabeth Warren.*"
Sander_files = "Bernie Sanders.*"
Harris_files = "Kamala Harris.*"
ORourke_files = "Beto O'Rourke.*"
Gillibrand_files = "Kirsten Gillibrand.*"
Klobucha_files = "Amy Klobucha.*"
Booker_files = "Cory Booker.*"
Castro_files = "JuliÃ¡n Castro.*"

Trump_data = PlaintextCorpusReader(tweets_root,Trump_files)
Warren_data = PlaintextCorpusReader(tweets_root,Warren_files)
Sander_data = PlaintextCorpusReader(tweets_root,Sander_files)
Harris_data = PlaintextCorpusReader(tweets_root,Harris_files)
ORourke_data = PlaintextCorpusReader(tweets_root,ORourke_files)
Gillibrand_data = PlaintextCorpusReader(tweets_root,Gillibrand_files)
Klobucha_data = PlaintextCorpusReader(tweets_root,Klobucha_files)
Booker_data = PlaintextCorpusReader(tweets_root,Booker_files)
Castro_data = PlaintextCorpusReader(tweets_root,Castro_files)

comparative_emotion_analyzer([
    ['Trump',Trump_data.raw()],
    ['Warren',Warren_data.raw()],
    ['Sander',Sander_data.raw()],
    ['Harris',Harris_data.raw()],
    ['ORourke',ORourke_data.raw()],
    ['Gillibrand',Gillibrand_data.raw()],
    ['Booker',Booker_data.raw()],
    ['Castro',Castro_data.raw()]],
    object_name='Condidate')
```

```python
sentiment_analysis2 = comparative_emotion_analyzer([
    ['Trump',Trump_data.raw()],
    ['Warren',Warren_data.raw()],
    ['Sander',Sander_data.raw()],
    ['Harris',Harris_data.raw()],
    ['ORourke',ORourke_data.raw()],
    ['Gillibrand',Gillibrand_data.raw()],
    ['Booker',Booker_data.raw()],
    ['Castro',Castro_data.raw()]],
    object_name='Condidate')
```

```python
for i in sentiment_analysis2.columns:
    sentiment_analysis2[i] = sentiment_analysis2[i].apply(lambda x: x*100)
```
